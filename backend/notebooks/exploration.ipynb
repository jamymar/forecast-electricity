```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/consumption_data_cleaned.csv")

df['start_date'] = pd.to_datetime(df['start_date'], utc=True)
df['hour'] = df['start_date'].dt.hour
df['day_of_week'] = df['start_date'].dt.dayofweek
df['month'] = df['start_date'].dt.month
```

```python
df.groupby('hour')['avg_value_hourly'].mean().plot(kind='bar', figsize=(12, 4), title='Conso moyenne par heure')
plt.tight_layout()
```

```python
df.groupby('day_of_week')['avg_value_hourly'].mean().plot(kind='bar', figsize=(12, 4), title='Conso moyenne par jour de la semaine (0=lundi)')
plt.tight_layout()
```

```python
df.groupby('month')['avg_value_hourly'].mean().plot(kind='bar', figsize=(12, 4), title='Conso moyenne par mois')
plt.tight_layout()
```
